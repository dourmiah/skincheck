```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL de l'image
url = "https://skincheck-bucket.s3.eu-west-3.amazonaws.com/skincheck-dataset/train/Acne+and+Rosacea+Photos/07Acne081101.jpg"

# Tlcharger l'image
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Afficher l'image
plt.imshow(img)
plt.axis('off')  # Pour ne pas afficher les axes
plt.show()
```
